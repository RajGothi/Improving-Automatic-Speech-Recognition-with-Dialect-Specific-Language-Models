```python
from transformers import Wav2Vec2ForCTC,Wav2Vec2ProcessorWithLM
import torch
import librosa
import IPython
```

```python
#Bhojpuri Language ex., Similarly for Bengali language by passing trained Bengali model.
model = Wav2Vec2ForCTC.from_pretrained("../Trained_Model/wav2vec2-bh")
processor = Wav2Vec2ProcessorWithLM.from_pretrained("../Trained_Model/wav2vec2-bh") #with LM
```

```python
audio_path = "Sample/281474978955505.wav"
IPython.display.Audio(audio_path)
```

```python
speech, sr = librosa.load(audio_path, sr=processor.feature_extractor.sampling_rate) 
input_values = processor(speech, return_tensors="pt",sampling_rate=16000).input_values

with torch.no_grad():
    logits = model(input_values).logits

transcription = processor.decode(logits[0].numpy())[0]
transcription = ''.join(transcription)
transcription
```
