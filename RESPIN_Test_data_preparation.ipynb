```python
from datasets import Dataset, DatasetDict

def convert_list_to_dataset_HF(dataset):
    data_dict = {
    "id": [item[0] for item in dataset],
    "path" : [item[1] for item in dataset],
    "lang_id" : [item[2] for item in dataset]
    }
    return Dataset.from_dict(data_dict)
dev_data_hf = convert_list_to_dataset_HF(selected_dev_dataset)
```

```python
from datasets import Audio
dev_dataset = dev_data_hf.cast_column("path", Audio(sampling_rate=16000))
```

```python
def prepare_dataset(batch):
    audio = batch["path"]

    # batched output is "un-batched" to ensure mapping is correct
    batch["input_values"] = processor(audio["array"], sampling_rate=audio["sampling_rate"]).input_values[0]
    batch["input_length"] = len(batch["input_values"])
    
    with processor.as_target_processor():
        batch["lang_id"] = batch["lang_id"] 
    return batch
```

```python
from transformers import Wav2Vec2Processor
processor = Wav2Vec2Processor.from_pretrained("bhojpuri_1")
dev_dataset = dev_dataset.map(prepare_dataset,remove_columns=['path'] ,num_proc=4)
```
